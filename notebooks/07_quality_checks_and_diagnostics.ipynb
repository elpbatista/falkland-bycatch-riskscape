# Quality Checks and Diagnostics

Quality checks keep the workflow honest. This section highlights the inspections needed to catch broken joins, date-key problems, unexpected missingness, implausible feature ranges, weak validation designs, and plots that no longer match the intended products.


## Data and Feature Checks

Preprocessing checks should verify required columns, consistent `h3` and `date` keys, duplicate records, missing values, expected dtypes, and row counts after joins. Derived variables such as gradients and anomalies should be checked for row inflation and plausible ranges.

## Model Diagnostics

Diagnostics include learner screening, structured validation, feature importance, partial dependence, observed-vs-predicted comparisons, and plausibility sensitivity. Production-fit diagnostics are retained for reproducibility and inspection but are not independent validation because the production model is fit to all balanced training rows.

## Known Limitations to Track

- Telemetry records are limited in time and individuals, especially for BBAL.
- Risk scores are not calibrated against independent bycatch observer records.
- Fishing exposure is currently generic in the core risk equation, although gear fields are retained for operational examples.
- The plausibility-gate parameter is a demonstration value rather than independently estimated.
- Seascape-conditioned products are exploratory summaries, not replacements for full H3 prediction surfaces.
